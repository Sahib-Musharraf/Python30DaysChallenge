Object-oriented programming (OOP) is a programming paradigm conpets that is based on the concept of objects, which can contain data and code that manipulates that data. Python's OOPs allows us to organize our code into reusable, modular, and efficient structures.



Well walk through the concepts of Class, Object, Attributes, and Methods using a Car example:


1. Class  The Structure

A class represents a general idea of an object. For example, a Car class defines what every car should know and do.

```python
class Car:
    pass
```

2. Object  A real world example

An object is an actual item built using the class. It has its own data.


```python
my_car = Car() #Here, my_car is a real car created from the Car class blueprint.
```

3. Attributes  Details of the Car

Attributes store information about each specific car, such as the manufacturer, engine type, seating capacity, and fuel type.

```python
class Car:
    def __init__(self, manufacturer, engine_type, seats, fuel_type):
        self.manufacturer = manufacturer
        self.engine_type = engine_type
        self.seats = seats
        self.fuel_type = fuel_type
```

4. Methods  What the Car Can Do

Methods are actions the car can perform. For now, let's make one to show its details.

```python
    def show_specs(self):
        print(f"Manufacturer: {self.manufacturer}")
        print(f"Engine Type: {self.engine_type}")
        print(f"Seating Capacity: {self.seats}")
        print(f"Fuel Type: {self.fuel_type}")
```

 Challenge
- Create a Car class with attributes and a display method

```python
class Car:
    def __init__(self, manufacturer, engine_type, seats, fuel_type):
        self.manufacturer = manufacturer
        self.engine_type = engine_type
        self.seats = seats
        self.fuel_type = fuel_type

    def show_specs(self):
        print(f"Manufacturer: {self.manufacturer}")
        print(f"Engine Type: {self.engine_type}")
        print(f"Seating Capacity: {self.seats}")
        print(f"Fuel Type: {self.fuel_type}")

# Create car objects
car1 = Car("Mercedes-Benz", "PL - 300", 5, "Petrol")
car2 = Car("Hyundai", "Electric Motor", 7, "Electric")

# Display their specs
car1.show_specs()
print()  # spacer
car2.show_specs()
```
