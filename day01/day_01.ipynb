‚úÖ What is Syntax in Python?

Let's understand what is Syntax in Python.....

Syntax refers to the set of rules that defines how a Python program is written and interpreted.
Think of it like grammar in a spoken language just as English has rules (like subject-verb agreement), 
Python has rules for how code must be written so the computer can understand and run it.

Simple Analogy:
English sentence: I am eating.

Python statement: print("Hello, world!")

Both must follow correct structure (syntax) to make sense.

üêç Examples of Python Syntax:
1. Printing something = print("Hello, world!")

2. Variables: x = 10

3. Indentation
Python uses indentation (spaces/tabs) to define blocks of code.
Ex: if x > 5:
    print("x is greater than 5")

4. Comments  
Comment is a line of text that the Python interpreter ignores when running the code. Comments are used to explain the code, make notes, or temporarily disable parts of the code during debugging.
Ex: # This is a single-line comment

     """

      multi-line comment
     """
